# Making plots

```python
import matplotlib.pyplot
import numpy
```

```python
arthritis = numpy.loadtxt(fname='data/inflammation-01.csv', delimiter=',')
```

```python
image = matplotlib.pyplot.imshow(arthritis)
matplotlib.pyplot.show()
```

```python
average_inflammation = arthritis.mean(axis=0)
avarage_plot = matplotlib.pyplot.plot(average_inflammation)
matplotlib.pyplot.show()
```

```python
std_inflammation_plot = matplotlib.pyplot.plot(arthritis.std(axis=0))
matplotlib.pyplot.show()
```

## Subplotting

```python
import numpy as np
import matplotlib.pyplot as plt

arthritis = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

figure = plt.figure(figsize=(10, 3))

axes1 = figure.add_subplot(1, 3, 1) # 1 row, 3 columns, 1st graph
axes2 = figure.add_subplot(1, 3, 2)
axes3 = figure.add_subplot(1, 3, 3)

axes1.set_ylabel('average')
axes1.plot(arthritis.mean(axis=0))

axes2.set_ylabel('maximum')
axes2.plot(arthritis.max(axis=0))

axes3.set_ylabel('minimum')
axes3.plot(arthritis.min(axis=0))

figure.tight_layout()
plt.show()
```
